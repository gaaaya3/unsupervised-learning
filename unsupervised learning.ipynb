```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
print(data.head())
```

```python
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
k = 3 
```

```python
kmeans = KMeans(n_clusters=k, random_state=42)
```

```python
data['KMeans_Cluster'] = kmeans.fit_predict(data)
```

```python
plt.figure(figsize=(10, 6))
```

```python
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='KMeans_Cluster', palette='viridis', data=data)
plt.title('KMeans Clustering on Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Cluster')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
hierarchical = AgglomerativeClustering(n_clusters=k)
```

```python
data['Hierarchical_Cluster'] = hierarchical.fit_predict(data)
```

```python
plt.figure(figsize=(12, 8))
```

```python
linked = linkage(data[iris.feature_names], method='ward')
```

```python
dendrogram(linked, orientation='top', labels=data.index.tolist(), distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
```

```python
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='Hierarchical_Cluster', palette='viridis', data=data)
plt.title('Hierarchical Clustering on Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Cluster')
plt.show()
```
